```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def membership_A(x):
    if x < 30:
        return 1
    elif 30 <= x < 50:
        return (50 - x) / (50 - 30)
    elif 50 <= x < 70:
        return (70 - x) / (70 - 50)
    else:
        return 0
```

```python

x=np.arange(21,121,10)
y=np.array([membership_A(i) for i in x])
plt.title("speed vs Low Speed Limit")
plt.plot(x,y)
print(membership_A(40))
```

```python
def membership_B(x):
    if x < 60:
        return 0
    elif 60 <= x < 80:
        return (x - 60) / (80 - 60)
    elif 80 <= x < 100:
        return (x - 80) / (100 - 80)
    else:
        return 1
```

```python
plt.title("Speed vs High Speed Limit")
y=np.array([membership_B(i) for i in x])
plt.plot(x,y)
```

```python
x_values = np.linspace(0, 120, 500)
A_values = np.array([membership_A(x) for x in x_values])
B_values = np.array([membership_B(x) for x in x_values])
```

```python
# i. Union
union_values = np.maximum(A_values, B_values)
```

```python
# ii. Intersection
intersection_values = np.minimum(A_values, B_values)
```

```python
complement_A_values = 1 - A_values
complement_B_values = 1 - B_values
```

```python
# iv. Difference
difference_values = A_values - B_values
```

```python
# Plotting the results
plt.figure(figsize=(14, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x_values, union_values, label='A  B')
plt.title('Union (A  B)')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Value')
plt.legend()
```

```python
plt.subplot(2, 2, 2)
plt.plot(x_values, intersection_values, label='A  B')
plt.title('Intersection (A  B)')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Value')
plt.legend()
```

```python
plt.subplot(2, 2, 3)
plt.plot(x_values, complement_A_values, label='A\'')
plt.plot(x_values, complement_B_values, label='B\'')
plt.title('Complement of A and B')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Value')
plt.legend()
```

```python
plt.subplot(2, 2, 4)
plt.plot(x_values, difference_values, label='A - B')
plt.title('Difference (A - B)')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Value')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def membership_C(x):
    if x < -10:
        return 0
    elif -10 <= x < 0:
        return (x + 10) / 10
    elif 0 <= x <= 5:
        return 1
    else:
        return 0
```

```python
def membership_D(x):
    if x < 25:
        return 0
    elif 25 <= x < 35:
        return (x - 25) / 10
    elif 35 <= x <= 40:
        return 1
    else:
        return 0
```

```python
# Trapezoidal Membership Function
def trapezoidal(x, a, b, c, d):
    return max(min((x-a)/(b-a), 1, (d-x)/(d-c)), 0)

# Triangular Membership Function
def triangular(x, a, b, c):
    return max(min((x-a)/(b-a), (c-x)/(c-b)), 0)

# Gaussian Membership Function
def gaussian(x, c, sigma):
    return np.exp(-0.5 * ((x - c) / sigma) ** 2)
```

```python
# Example Fuzzy Rules
rules = [
    ("Risky", "High"),
    ("Average", "Moderate"),
    ("Excellent", "Slow")
]
```

```python
def risky_temperature(x):
    return max(min((x - 70) / 20, 1, (100 - x) / 20), 0)

def average_temperature(x):
    return max(min((x - 40) / 20, 1, (60 - x) / 20), 0)

def excellent_temperature(x):
    return max(min((x - 10) / 20, 1, (30 - x) / 20), 0)
```

```python
def slow_speed(y):
    return max(min((y - 50) / 100, 1, (150 - y) / 100), 0)

def moderate_speed(y):
    return max(min((y - 150) / 100, 1, (250 - y) / 100), 0)

def high_speed(y):
    return max(min((y - 250) / 100, 1, (350 - y) / 100), 0)
```

```python
temperature_funcs = {
    "Risky": risky_temperature,
    "Average": average_temperature,
    "Excellent": excellent_temperature
}
```

```python
speed_funcs = {
    "Slow": slow_speed,
    "Moderate": moderate_speed,
    "High": high_speed
}
```

```python
import numpy as np
def mamdani_aggregation(temp, temperature_funcs, rules):
    aggregated_membership = np.zeros_like(x_values_num)
    for rule in rules:
        temp_membership = temperature_funcs[rule[0]](temp)
        speed_membership = np.minimum(temp_membership, np.vectorize(speed_funcs[rule[1]])(x_values_num))
        aggregated_membership = np.maximum(aggregated_membership, speed_membership)
    return aggregated_membership
```

```python
temp = 85
aggregated_membership = mamdani_aggregation(temp, temperature_funcs, rules)
```

```python
plt.plot(x_values_num, aggregated_membership, label=f'Aggregated Membership at Temp={temp}')
plt.title('Aggregated Fuzzy Output for Fan Speed')
plt.xlabel('Fan Speed (RPM)')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
def defuzzify(aggregated_membership, x_values):
    # Handle potential division by zero
    if np.sum(aggregated_membership) == 0:
        return 0  # Or any other default value you deem appropriate
    else:
        return np.sum(aggregated_membership * x_values) / np.sum(aggregated_membership)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example membership function (aggregated membership)
x_values_num = np.linspace(0, 1000, 100)  # Example range of fan speeds
aggregated_membership = np.maximum(0, 1 - np.abs(x_values_num - 500) / 200)  # Example triangular membership




# Centroid Defuzzification function
def centroid_defuzzification(aggregated_membership, x_values):
    return np.sum(aggregated_membership * x_values) / np.sum(aggregated_membership)

# Calculate the defuzzified fan speed
fan_speed = centroid_defuzzification(aggregated_membership, x_values_num)
print(f"Defuzzified Fan Speed: {fan_speed:.2f} RPM")

# Plot the aggregated membership function
plt.figure(figsize=(10, 6))
plt.plot(x_values_num, aggregated_membership, label='Aggregated Membership Function', color='blue')

# Mark the defuzzified fan speed on the plot
plt.axvline(x=fan_speed, color='red', linestyle='--', label=f'Defuzzified Fan Speed: {fan_speed:.2f} RPM')
plt.fill_between(x_values_num, 0, aggregated_membership, color='blue', alpha=0.3)

# Labeling the plot
plt.title('Centroid Defuzzification of Fan Speed')
plt.xlabel('Fan Speed (RPM)')
plt.ylabel('Membership Value')
plt.legend()

# Show the plot
plt.show()
```
